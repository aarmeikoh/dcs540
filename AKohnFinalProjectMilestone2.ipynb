```python
import pandas as pd
import numpy as np
```

```python
csv_data = pd.read_csv('2005_-_2020_Quality_Review_Ratings.csv') # Read data set from csv
csv_data.head()
```

```python
csv_data['Borough'] = csv_data['BN'].apply(lambda x: x[0]) # Extract borough code (may require further wrangling to match with other data sets)
csv_data['School_Num'] = csv_data['BN'].apply(lambda x: x[1:]) # Extract school number to be matched with other data sets
csv_data.replace('No Data', np.nan, inplace= True) # Switch missing data values to NaN

csv_data.tail()
```

```python
csv_data['Indicator_1.2'].unique()
```

```python
# Replace rating categories with orderd numbers
key = {'U' : 1, # U/UD = underdeveloped
      'UD' : 1,
      'UPF' : 1.5, # UPF = underdeveloped with proficient features
      'D' : 2, # D = developing
      'P' : 3, # P = proficient
      'WD' : 4, # WD = well developed
      'O' : 5}  # O = outstanding
csv_data.replace(key, inplace= True) # Use replace function to replace values
csv_data.head(6)
```

```python
csv_data['Borough'].unique() # Find all unique Borough letter codes
```

```python
csv_data['Borough'] = csv_data['Borough'].apply(lambda x : x.upper()) # Convert all to upper case
csv_data['Borough'].unique()
```
